```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt 
from pyMCDS import pyMCDS
```

```python
type1_matrix = np.zeros((3,10))
type2_matrix = np.zeros((3,10))
time_matrix  = np.zeros((3,10))
direc_lst    = ('C:\\Users\zeXal\Desktop\stem_cells_paul\stem_cells-master\output_fix',
                'C:\\Users\zeXal\Desktop\stem_cells_paul\stem_cells-master\output_r1',
                'C:\\Users\zeXal\Desktop\stem_cells_paul\stem_cells-master\output')
```

```python
direc_lst[1]
```

```python
for j in range(3):
    if j == 0:
        k = 10
    else:
        k =6        
    for i in range(k):
        if i>=10:
            new_str = str(i);
        else:
            new_str = '0'+str(i)
        my_str = 'output000000'+new_str +'.xml'
        mcds = pyMCDS(my_str,direc_lst[j])
        all_type = mcds.data['discrete_cells']['cell_type']
        type1_matrix[j][i] = sum(all_type==1)
        type2_matrix[j][i] = sum(all_type==2)
        time_matrix[j][i]  = mcds.get_time()
```

```python
plt.figure(figsize=(16,5))
title_lst = ('fix duration for division 1/6h','stocastic duration for division one 1/6h','stocastic duration for division two 1/6h ')
for i in range(3):
    if i ==0:
        l = -1
    else:
        l = -4
    plt.subplot(1,3,i+1)
    plt.plot(time_matrix[i,:l],type1_matrix[i,:l],'ro-',label='type1 number')
    plt.plot(time_matrix[i,:l],type2_matrix[i,:l],'bo-',label='type2 number')
    plt.legend()
    plt.title(title_lst[i])
    plt.xlim(0, 3240)
```
